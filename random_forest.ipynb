```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
from sklearn.tree import plot_tree
```

```python
dir(digits)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns='target'), df.target, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=110)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
print(model.estimators_[0])
```

```python
individual_tree = model.estimators_[1]

plt.figure(figsize=(12, 8))
plot_tree(individual_tree, filled=True)
plt.show()
```
